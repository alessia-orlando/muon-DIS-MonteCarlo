```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Read the csv file and print the first 5 rows of the dataset:

```python
df = pd.read_csv("data.csv")
print(df.head())
```

With Pandas, make a histogram of the data frame's columns:

```python
df.hist(bins=30, figsize=(15, 10))
plt.suptitle("Data Frame histogram")
plt.tight_layout()
plt.show()
```

Focusing on Starting Station:

```python
# Count the frequency of each Starting station
ss_counts = df['Starting station'].value_counts().sort_index()

# Plot
plt.figure()
ss_counts.plot(kind='bar')
plt.title("Starting station distribution")
plt.xlabel("Starting station")
plt.ylabel("Entries")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

To visualize how the feature values vary across different classes (Starting station), plot their mean value for each Starting Station:

```python
# Mean value
mean_by_station = df.groupby("Starting station").mean(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(mean_by_station, annot=True, fmt=".2f")
plt.title("Features mean value for each Starting station")
plt.xlabel("Feature")
plt.ylabel("Starting station")
plt.tight_layout()
plt.show()
```

To show the relationship between the features, plot the correlation matrix (+1 = positive correlation, 0 = no correlation, -1 = negative correlation):

```python
plt.figure(figsize=(14, 8))
sns.heatmap(df.drop(columns='Starting station').corr(), annot=True, fmt=".2f")
plt.title("Correlation matrix")
plt.show()
```

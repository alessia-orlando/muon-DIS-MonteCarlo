# Dataset exploration

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset, show the shape and the first 5 rows.

```python
# Load
df = pd.read_csv("data.csv")

# Shape and head
print(f"Shape: {df.shape}")
print(df.head())
```

## Dataset Infos

Check data types, non-null counts, and missing values.

```python
# Data types and non-null counts
df.info()

# Count of missing values
df.isnull().sum()
```

## Distributions

With pandas, visualize the distributions of each column.

```python
df.hist(bins=30, figsize=(15, 10))
plt.suptitle("Data Frame histograms")
plt.tight_layout()
plt.show()
```

Focusing on Starting Station:

```python
# Count the frequency of each Starting station
ss_counts = df['Starting station'].value_counts().sort_index()

# Plot
plt.figure()
ss_counts.plot(kind='bar')
plt.title("Starting station distribution")
plt.xlabel("Starting station")
plt.ylabel("Entries")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 

To visualize how the features (number of hits, QCD/hits, and shower width for each Starting station) values vary across different classes (Starting station), plot their mean value per Starting Station.

```python
# Mean value
mean_by_station = df.groupby("Starting station").mean(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(mean_by_station, annot=True, fmt=".2f")
plt.title("Features mean value for each Starting station")
plt.xlabel("Feature")
plt.ylabel("Starting station")
plt.tight_layout()
plt.show()
```

## Correlation Matrix

To show the relationship between the features, plot the correlation matrix (+1 = positive correlation, 0 = no correlation, -1 = negative correlation):

```python
plt.figure(figsize=(14, 8))
sns.heatmap(df.drop(columns='Starting station').corr(), annot=True, fmt=".2f")
plt.title("Correlation matrix")
plt.show()
```
